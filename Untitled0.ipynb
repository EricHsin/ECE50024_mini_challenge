```python
import numpy as np
import cv2
import os
from PIL import Image

def cropImg(imgPath):
    faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    try:
        image = Image.open(imgPath).convert('RGB')
    except Exception as e:
        print(f"PIL error: {e}")
        return None

    # Convert PIL Image to numpy array for OpenCV to process
    open_cv_image = np.array(image)
    # Convert RGB to BGR for OpenCV
    open_cv_image = open_cv_image[:, :, ::-1].copy()

    gray = cv2.cvtColor(open_cv_image, cv2.COLOR_BGR2GRAY)
    faces = faceCascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    if len(faces) > 0:
        x, y, w, h = faces[0]
        face = open_cv_image[y:y+h, x:x+w]
        return face  # Return numpy array in BGR format
    else:
        return open_cv_image  # Return the original image in BGR format as a numpy array

image_folder = 'train'
output_folder = 'processed'
count = 0

# Ensure the output directory exists
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

for filename in os.listdir(image_folder):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg')):  # Check for image files
        file_path = os.path.join(image_folder, filename)

        croppedFace = cropImg(file_path)
        count += 1
        if(count % 100 == 1):
          print(count)
        if croppedFace is not None:
            croppedImgPath = os.path.join(output_folder, filename)
            cv2.imwrite(croppedImgPath, croppedFace)
        else:
            print(f"Skipping file due to loading error: {filename}")

print("Processing complete.")
```

```python
from torch.utils.data import Dataset
import pandas as pd
import os
from PIL import Image
import torch
from torchvision import transforms

class CustomDataset(Dataset):
    def __init__(self, image_folder, train_csv_file, category_csv_file, transform=None):
        self.image_folder = image_folder
        self.transform = transform
        self.train_df = pd.read_csv(train_csv_file)
        self.train_df['exists'] = self.train_df['File Name'].apply(lambda x: os.path.exists(os.path.join(image_folder, x)))
        self.train_df = self.train_df[self.train_df['exists']]
        self.train_df['Label'] = self.train_df['Category'].map(category_to_index)
        self.filenames = self.train_df['File Name'].tolist()
        self.labels = self.train_df['Label'].tolist()

    def __len__(self):
        return len(self.filenames)

    def __getitem__(self, idx):
        img_path = os.path.join(self.image_folder, self.filenames[idx])
        image = Image.open(img_path).convert('RGB')

        if self.transform:
            image = self.transform(image)

        label = self.labels[idx]
        label_tensor = torch.tensor(label, dtype=torch.long)

        return image, label_tensor



# Example usage
image_folder = '/content/content/processed'
label_csv_file = '/content/train.csv'
category_file = '/content/category.csv'

transformer = transforms.Compose([
    # Apply any desired data augmentation or preprocessing here
    transforms.Resize((224, 224), interpolation=transforms.InterpolationMode.BICUBIC),  # Example
    transforms.CenterCrop(224),
    transforms.ToTensor(),  # This converts PIL Images to Tensors
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),  # Example normalization
])

dataset = CustomDataset(image_folder, label_csv_file, category_file, transformer)

print(dataset.labels[3000], dataset.filenames[3000])
```

```python
from torch.utils.data import random_split

total_size = len(dataset)
train_size = int(total_size)
validate_size = total_size - train_size

train_dataset, validate_dataset = random_split(dataset, [train_size, validate_size])
```

```python
batchSize = 32
trainLoader = DataLoader(train_dataset, batch_size = batchSize, shuffle = True, num_workers = 4, drop_last=False, pin_memory=True)
valLoader = DataLoader(validate_dataset, batch_size = batchSize, shuffle = False, num_workers = 4, drop_last=False, pin_memory=True)
```

```python
transformer = transforms.Compose([
    transforms.Resize((224, 224), interpolation=transforms.InterpolationMode.BICUBIC),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

class TestDataset(Dataset):
    def __init__(self, image_folder, transform=None):
        self.image_folder = image_folder
        self.transform = transform
        self.filenames = sorted([f for f in os.listdir(image_folder) if os.path.isfile(os.path.join(image_folder, f))])
        self.faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

    def __len__(self):
        return len(self.filenames)

    def __getitem__(self, idx):
        img_path = os.path.join(self.image_folder, self.filenames[idx])
        image = Image.open(img_path).convert('RGB')
        open_cv_image = np.array(image)
        open_cv_image = open_cv_image[:, :, ::-1].copy()

        gray = cv2.cvtColor(open_cv_image, cv2.COLOR_BGR2GRAY)
        faces = self.faceCascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

        if len(faces) > 0:
            x, y, w, h = faces[0]
            face = open_cv_image[y:y+h, x:x+w]
            face = Image.fromarray(face[:, :, ::-1])
        else:
            face = image

        if self.transform:
            face = self.transform(face)

        return face, self.filenames[idx]
```

```python
test_image_folder = '/content/test'
test_dataset = TestDataset(test_image_folder, transformer)
test_loader = DataLoader(test_dataset, batch_size=1, shuffle=False)
```

```python
model.eval()  # Set the model to evaluation mode
predictions = []
filenames = []

with torch.no_grad():
    for images, fns in test_loader:
        images = images.to(device)

        outputs = model(images)
        _, pred_indices = torch.max(outputs, 1)

        pred_labels = [index_to_category[idx.item()] for idx in pred_indices]

        predictions.extend(pred_labels)
        filenames.extend(fns)


```
